```python
import plotly.express as px
```

```python
import numpy as np
```

```python
x=np.random.randint(1,300,size=100)
y=np.random.randint(1,300,size=100)
```

```python
x,y
```

```python
fig=px.scatter(x=x,y=y,labels={"x":"nandha","y":"kumar"},title="Expense")
fig.update_layout(title_x=0.5,title_y=0.5)
fig.show()
```

```python
fig=px.bar(x=x,y=y,labels={"x":"nandha","y":"kumar"},title="Expense")
fig.update_layout(title_x=0.5,title_y=0.5)
fig.show()
```

```python
fig=px.line(x=x,y=y,labels={"x":"nandha","y":"kumar"},title="Expense")
fig.update_layout(title_x=0.5,title_y=0.5)
fig.show()
```
