```python
#chi square test practice
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df
```

```python
data=pd.crosstab(df["smoker"],df["sex"])
data
```

```python
data=pd.crosstab(df["day"],df["size"])
data
```

```python
obs_value=data.values
obs_value
```

```python
from scipy import stats
```

```python
value=stats.chi2_contingency(data)
value
```

```python
exp_value=value[3]
exp_value
```

```python
chi_square=sum(sum([((obs_value-exp_value)**2)/exp_value for obs_value,exp_value in zip(obs_value,exp_value)]))
chi_square
```

```python
#level of significance
# p=0.075
# p+q=1   ===>q=1-0.075
crt_value=stats.chi2.ppf(q=.925,df=15)
crt_value
```

```python
if chi_square >= crt_value:
  print("ho is rejected,ha is accepted,there is a relationship between(size & day)")
else:
  print("ho is accepted,ha is rejected,there is no relationship between(size & day)")
```

```python
p_value=1-stats.chi2.cdf(x=chi_square,df=15)
p_value
```

```python
if p_value < 0.075:
  print("ho is rejected,ha is accepted,there is a relationship between(size & day)")
else:
  print("ho is accepted,ha is rejected,there is no relationship between(size & day)")
```
